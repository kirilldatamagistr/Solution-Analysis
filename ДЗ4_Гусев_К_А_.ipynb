```python



#                                ___2.1___




import pandas as pd

#---------------------------------------------------------------------------------------------------------------------------

# users
users['churn_date'] = pd.to_datetime(users['churn_date'])
users['reg_date'] = pd.to_datetime(users['reg_date'])
users['churn_date'] = users['churn_date'].fillna(users['reg_date'].max())

# calls
calls['call_date'] = pd.to_datetime(calls['call_date'])
calls['month'] = pd.DatetimeIndex(calls['call_date']).month

def cf(a):
    return math.ceil(a)

calls['duration'] = calls['duration'].apply(cf)

# messages
messages['message_date'] = pd.to_datetime(messages['message_date'])
messages['month'] = pd.DatetimeIndex(messages['message_date']).month

# internet
internet['session_date'] = pd.to_datetime(internet['session_date'])
internet['month'] = pd.DatetimeIndex(internet['session_date']).month

#---------------------------------------------------------------------------------------------------------------------------

t_calls = calls.pivot_table(index=['user_id', 'month'], values='duration', aggfunc='sum').reset_index()
t_internet = internet.pivot_table(index=['user_id', 'month'], values='mb_used', aggfunc='sum').reset_index()
t_messages = messages.pivot_table(index=['user_id', 'month'], values='message_date', aggfunc='count').reset_index()
t = t_calls.merge(t_internet, how='outer', on=['user_id', 'month']).merge(t_messages, how='outer', on=['user_id', 'month'])
t['duration'] = t['duration'].fillna(0).astype('int')
t['mb_used'] = t['mb_used'].fillna(0)
t['message'] = t['message_date'].fillna(0).astype('int')
t = t.drop('message_date', axis=1)
tariffs = tariffs.rename({'tariff_name' : 'tariff'}, axis=1)
t = t.merge(users, how='left', on='user_id').merge(tariffs, how='left', on='tariff')

#---------------------------------------------------------------------------------------------------------------------------
```

```python



#                                ___2.2___




def prepare_data(users, calls, messages, internet):

    print("Проверка пользователей:")
    print(("=" * 45))
    print(f"Дубликаты user_id: {users['user_id'].duplicated().sum()}")
    print(f"Пропуски в возрасте: {users['age'].isna().sum()}")
    print(f"Некорректный возраст: {(users['age'] < 0).sum()}")
    print(("=" * 45))
    users = users[users['age'] > 0] # Удаляет пользователей с отрицательным возрастом
    print("Проверка звонков:")
    print(f"Звонки с нулевой длительностью: {(calls['duration'] == 0).sum()}")
    print(f"Отрицательная длительность: {(calls['duration'] < 0).sum()}")
    print(("=" * 45))
    calls = calls[calls['duration'] > 0] # Удаляет звонки с нулевой и отрицательной длительностью

    print("Проверка сообщений:")
    print(f"Дубликаты id: {messages['id'].duplicated().sum()}")
    print(("=" * 45))
    print("Проверка интернет-сессий:")
    print(f"Отрицательный трафик: {(internet['mb_used'] < 0).sum()}")
    print(("=" * 45))
    internet = internet[internet['mb_used'] > 0] # Удаляет сессии с отрицательным трафиком

    print("Проверка согласованности:")
    all_user_ids = set(users['user_id'])
    calls_users = set(calls['user_id'])
    messages_users = set(messages['user_id'])
    internet_users = set(internet['user_id'])

    print(f"Пользователи в calls, но не в users: {len(calls_users - all_user_ids)}")
    print(f"Пользователи в messages, но не в users: {len(messages_users - all_user_ids)}")
    print(f"Пользователи в internet, но не в users: {len(internet_users - all_user_ids)}")
    print(("=" * 45))
    print('Данный код проверяет и если необходимо:')
    print('В users удаляет пользователей с отрицательным возрастом')
    print('В calls удаляет звонки с нулевой и отрицательной длительностью')
    print('В internet удаляет сессии с отрицательным трафиком')
    print('В calls, messages, internet удаляет записи о звонках, сообщениях и интернете для пользователей, которых нет в users')

    calls = calls[calls['user_id'].isin(all_user_ids)] # Удаляет записи о звонках, сообщениях и интернете для пользователей, которых нет в users
    messages = messages[messages['user_id'].isin(all_user_ids)] # Удаляет записи о звонках, сообщениях и интернете для пользователей, которых нет в users
    internet = internet[internet['user_id'].isin(all_user_ids)] # Удаляет записи о звонках, сообщениях и интернете для пользователей, которых нет в users
    return users, calls, messages, internet
users, calls, messages, internet = prepare_data(users, calls, messages, internet)
```

```python



#                               ___2.3___




import pandas as pd
import numpy as np
from datetime import datetime

def data_control():

    tariffs = {
        'Базовый': {
            'monthly_fee': 550,
            'minutes_included': 500,
            'messages_included': 50,
            'gb_included': 15,
            'rub_per_minute': 3,
            'rub_per_message': 3,
            'rub_per_gb': 200
        },
        'Смарт': {
            'monthly_fee': 1950,
            'minutes_included': 3000,
            'messages_included': 1000,
            'gb_included': 30,
            'rub_per_minute': 1,
            'rub_per_message': 1,
            'rub_per_gb': 150
        }
    }

    def get_year_month(date):
        return f"{date.year}-{date.month:02d}"

    # 1. Расчет по звонкам
    calls['year_month'] = calls['call_date'].apply(get_year_month)
    monthly_calls = calls.groupby(['user_id', 'year_month']).agg(
        calls_count=('id', 'count'),
        minutes_used=('duration', 'sum')
    ).reset_index()

    # 2. Расчет по сообщениям
    messages['year_month'] = messages['message_date'].apply(get_year_month)
    monthly_messages = messages.groupby(['user_id', 'year_month']).agg(
        messages_count=('id', 'count')
    ).reset_index()

    # 3. Расчет по интернет-трафику
    internet['year_month'] = internet['session_date'].apply(get_year_month)
    monthly_internet = internet.groupby(['user_id', 'year_month']).agg(
        mb_used=('mb_used', 'sum')
    ).reset_index()

    # Преобразование МБ в ГБ (с округлением вверх)
    monthly_internet['gb_used'] = np.ceil(monthly_internet['mb_used'] / 1024)

    # Объединение всех данных
    monthly_data = pd.merge(monthly_calls, monthly_messages, on=['user_id', 'year_month'], how='outer')
    monthly_data = pd.merge(monthly_data, monthly_internet[['user_id', 'year_month', 'gb_used']],
                            on=['user_id', 'year_month'], how='outer')

    # Заполнение пропущенных значений нулями
    monthly_data = monthly_data.fillna(0)

    # Добавление информации о тарифе пользователя
    monthly_data = pd.merge(monthly_data, users[['user_id', 'tariff']], on='user_id')

    # Функция для расчета выручки
    def calculate_revenue(row):
        tariff_name = row['tariff']

        # Расширенная нормализация названия тарифа
        tariff_name_lower = str(tariff_name).lower().strip()

        if tariff_name_lower in ['смарт', 'smart', 'см']:
            normalized_tariff = 'Смарт'
        elif tariff_name_lower in ['базовый', 'базовий', 'base', 'база', 'basic']:
            normalized_tariff = 'Базовый'
        else:
            # Если тариф не распознан, используем оригинальное название
            normalized_tariff = tariff_name
            print(f"Предупреждение: неизвестный тариф '{tariff_name}'")

        tariff = tariffs[normalized_tariff]

        # Расчет дополнительных минут
        extra_minutes = max(0, row['minutes_used'] - tariff['minutes_included'])
        minutes_revenue = extra_minutes * tariff['rub_per_minute']

        # Расчет дополнительных сообщений
        extra_messages = max(0, row['messages_count'] - tariff['messages_included'])
        messages_revenue = extra_messages * tariff['rub_per_message']

        # Расчет дополнительного интернет-трафика
        extra_gb = max(0, row['gb_used'] - tariff['gb_included'])
        internet_revenue = extra_gb * tariff['rub_per_gb']

        # Общая выручка
        total_revenue = tariff['monthly_fee'] + minutes_revenue + messages_revenue + internet_revenue

        return total_revenue

    # Расчет выручки для каждого пользователя по месяцам
    monthly_data['monthly_revenue'] = monthly_data.apply(calculate_revenue, axis=1)

    # Вывод результатов
    print("Результаты помесячного анализа:")
    print("=" * 80)
    print(f"Всего записей: {len(monthly_data)}")
    print(f"Уникальных пользователей: {monthly_data['user_id'].nunique()}")
    print(f"Период анализа: с {monthly_data['year_month'].min()} по {monthly_data['year_month'].max()}")
    print("=" * 80)

    print("Первые 10 строк результатов:")
    print(monthly_data.head(10))

    # Сводная статистика по выручке
    print("=" * 80)
    print("Сводная статистика по помесячной выручке:")
    print(f"Средняя выручка: {monthly_data['monthly_revenue'].mean():.2f} руб.")
    print(f"Медианная выручка: {monthly_data['monthly_revenue'].median():.2f} руб.")
    print(f"Минимальная выручка: {monthly_data['monthly_revenue'].min():.2f} руб.")
    print(f"Максимальная выручка: {monthly_data['monthly_revenue'].max():.2f} руб.")

    # Статистика по тарифам
    print("=" * 80)
    print("Статистика выручки по тарифам:")
    tariff_stats = monthly_data.groupby('tariff')['monthly_revenue'].agg(['mean', 'median', 'min', 'max', 'count'])
    print(tariff_stats)

    monthly_data.to_csv('monthly_user_analysis.csv', index=False)
    print("=" * 80)
data_control()

def over_calls(row):
    return max(row['duration'] - row['minutes_included'], 0)
def over_messages(row):
    return max(row['message'] - row['messages_included'], 0)
def over_internet(row):
    return max(row['mb_used'] - row['mb_per_month_included'], 0)
t['over_calls'] = t.apply(over_calls, axis=1)
t['over_messages'] = t.apply(over_messages, axis=1)
t['over_internet'] = t.apply(over_internet, axis=1)
t['total'] = t['rub_monthly_fee'] + t['over_calls'] * t['rub_per_minute']  + t['over_messages'] * t['rub_per_message'] + np.ceil(t['over_internet'] / 1024) * t['rub_per_gb'].astype('int')
```

```python



#                               ___4.1___




#   H0: средняя выручка пользователей тарифов «Базовый» и «Смарт» одинкаовы
#   H1: средняя выручка пользователей тарифов «Базовый» и «Смарт» различаются
```

```python
t_base_total = t_base.pivot_table(index='user_id', values='total', aggfunc='sum').reset_index()
t_smart_total = t_smart.pivot_table(index='user_id', values='total', aggfunc='sum').reset_index()
print(f"Дисперсия доходов - Тариф Базовый: {t_base_total['total'].var():.2f}")
print(f"Дисперсия доходов - Тариф Смарт: {t_smart_total['total'].var():.2f}")
```

```python
sns.displot(t_base_total['total'],color='orange')
plt.xlabel('Суммарная выручка, руб')
plt.ylabel('Количество абонентов, чел.')
plt.xlim(0, 35000)
plt.ylim(0, 70)
plt.title('Тариф "Базовый"')
plt.grid(linestyle='--', alpha=0.5)
plt.show()

sns.displot(t_smart_total['total'],color='orange')
plt.xlabel('Суммарная выручка, руб')
plt.ylabel('Количество абонентов, чел.')
plt.xlim(0, 35000)
plt.ylim(0, 70)
plt.title('Тариф "Смарт"')
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
results = st.ttest_ind(t_base_total['total'], t_smart_total['total'], equal_var=False)
print(f'p-value: {results.pvalue}')
alpha = 0.05

if results.pvalue < alpha:
    print('--- >>> Отвергаем нулевую гипотезу <<< ---')
else:
    print('--- >>> Нет оснований отвергнуть нулевую гипотезу о том что тарифы приносят одинаковую прибыль <<< ---')

print(f"Средняя выручка с абонентов тарифа Базовый: {t_base_total['total'].mean():.0f}")
print(f"Средняя выручка с абонентов тарифа Смарт: {t_smart_total['total'].mean():.0f}")
print(f"Разница в средних: {t_smart_total['total'].mean() - t_base_total['total'].mean():.0f} руб.")
print(f"Тариф Смарт приносит выручки на {(t_smart_total['total'].mean() / t_base_total['total'].mean() - 1) * 100:.1f}% больше")
```

```python
results = st.ttest_ind(t_Moscow_total['total'], t_other_total['total'], equal_var=False)
print(f'p-value: {results.pvalue}')
alpha = 0.05
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('--->>> Нет оснований отвергнуть нулевую гипотезу <<<---')
```
